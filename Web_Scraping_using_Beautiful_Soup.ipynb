```python
#Install the necessary libraries
import pandas as pd 
import requests
from bs4 import BeautifulSoup
```

```python
#Creating an excel file and adding columns,titlename to the sheet
import openpyxl
excel=openpyxl.Workbook()
sheet=excel.active
sheet.title='Top Rated Movies'
sheet.append(['Movie Rank','Movie Name','Year of Release','IMDB Rating'])
```

```python
#Scraping IMDB top 250 movie ratings and saving the content in the excel sheet
try:
  headers = {'Accept-Language': 'en-US,en;q=0.5'}
  URL=requests.get("https://www.imdb.com/chart/top/",headers=headers)
  URL.raise_for_status()
  soup= BeautifulSoup(URL.text,'html.parser')
  movies= soup.find('tbody',class_="lister-list").find_all("tr")
  for movie in movies:
    name=movie.find('td',class_="titleColumn").a.text
    rank=movie.find("td",class_="titleColumn").get_text(strip=True).split(".")[0]
    year=movie.find("td",class_="titleColumn").span.text.strip('()')
    rating=movie.find("td",class_="ratingColumn imdbRating").strong.text
    print(rank,name,year,rating)
    sheet.append([rank,name,year,rating])
except Exception as e:
  print(e)
excel.save('IMDB top 250 movie Ratings.xlsx')
```

```python
#Creating the excel sheet and adding the sheet name and the column names
import openpyxl
excel=openpyxl.Workbook()
sheet=excel.active
sheet.title='Joblistings Simplyhired'
sheet.append(['Job Title','Company','Location','Description','Estimated Salary'])
```

```python
#Webscraping Job listings in Simplyhired and saving the content in the excel sheet
import requests
from bs4 import BeautifulSoup
URL = requests.get("https://www.simplyhired.com/search?q=software+developer&l=Miami%2C+FL")
URL.raise_for_status()
pagelimit=10
for page in range(1, pagelimit+1):
  url = f"{URL}&pn={page}"
  soup = BeautifulSoup(URL.content, 'html.parser')
  jobs = soup.find_all('li',class_='css-0')
  for job in jobs:
    jobname = job.find('a', class_="chakra-button css-12bkbc3").text.strip()
    companyname = job.find('span', class_="css-lvyu5j").text.strip('')
    location = job.find('span', class_="css-1t92pv").text.strip('()')
    description = job.find('p', class_="chakra-text css-jhqp7z").text.strip()
    salary_elem = job.find('p', class_='chakra-text css-1ejkpji')
    salary = salary_elem.text.strip() if salary_elem else 'N/A'
    
    print('Job:', jobname)
    print('Company:', companyname)
    print('Location:', location)
    print('Description:', description)
    print('Estimated Salary:', salary)
    print('---------------------------------------------------------------------------------------------------------------------')
    sheet.append([jobname,companyname,location,description,salary])
excel.save('Job Postings Simplyhired.com.xlsx')
```
